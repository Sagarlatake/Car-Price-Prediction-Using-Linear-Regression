```python
df_corr = pd.DataFrame(data.corr())
df_corr
```

```python
# get y filtered

df_corr_y = df_corr['price']
df_corr_y
```

```python
df_corr_y = pd.DataFrame(df_corr_y)
df_corr_y.reset_index(inplace=True)
df_corr_y
```

```python
df_corr_y.rename(columns={'index': 'col_name'}, inplace=True)
df_corr_y.head()
```

```python
# select required column correkllation greater than 0.5
selected_col = []
for r in df_corr_y.iterrows():
    
    if np.abs(r[1]['price'])>0.5:
        print(r[1]['price'])
        selected_col.append(r[1]['col_name'])
```

```python
# selected columns for prediction are       
selected_col
```

```python
data_final = data[selected_col]
data_final.head()
```

```python
x = data_final.drop(columns=['price'])
y = data_final['price']

print(x.shape)
print(y.shape)
```

# data splitting

```python
 from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size=0.2)
```

```python
print(xtrain.shape)
print(ytrain.shape)
print(xtest.shape)
print(ytest.shape)
```

### Build Model

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(xtrain,ytrain)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
pred = lr.predict(xtest)
pred = pd.DataFrame(pred, columns=['price'])
pred
```

## Calculate Error

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
mse = mean_squared_error(ytest,pred)
mae = mean_absolute_error(ytest, pred)
r2 = r2_score(ytest, pred)

print('MSE:',mse)
print('MAE:',mae)
print('R2_score:',r2)
```
